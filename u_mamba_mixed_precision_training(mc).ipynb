```python
"""
Enhanced ACDC Dataset Preprocessing Script with Data Augmentation

This script preprocesses the ACDC dataset for 3D segmentation tasks with advanced data augmentation.
It loads HDF5 files, processes them, applies augmentations, and saves as numpy arrays for training.
"""

import numpy as np
import h5py
from glob import glob
import os
import shutil
import re
import time
from sklearn.model_selection import train_test_split
from collections import namedtuple
import skimage.morphology as morph
import skimage.transform
import skimage.draw
from skimage.feature import canny
import cv2
from cv2 import bilateralFilter
from skimage.restoration import denoise_tv_chambolle, denoise_tv_bregman
import SimpleITK as sitk
import scipy.ndimage as snd

# Configuration for ACDC dataset
ACDC_SHAPE = (10, 224, 224)  # Depth, Height, Width for ACDC
rng = np.random.RandomState(40)

def mkdir(path):
    """Create a directory if it doesn't exist"""
    path = path.strip().rstrip("\\")
    if not os.path.exists(path):
        os.makedirs(path)
        print(f"Created directory: {path}")
    return path

# ********************** Weight Map Generation *******************#
selem = morph.disk(1)
def getEdgeEnhancedWeightMap(label, label_ids=[0,1,2,3], scale=1, edgescale=1, assign_equal_wt=False):
    """Generate edge-enhanced weight maps for segmentation"""
    shape = (0,) + label.shape[1:]
    weight_map = np.empty(shape, dtype='uint8')
    if assign_equal_wt:
        return np.ones_like(label)
    for i in range(label.shape[0]):
        # Estimate weight maps:
        weights = np.ones(len(label_ids))
        slice_map = np.ones(label[i,:,:].shape)
        for _id in label_ids:
            class_frequency = np.sum(label[i,:,:] == label_ids[_id])
            if class_frequency:
                weights[label_ids.index(_id)] = scale * label[i,:,:].size / class_frequency
                slice_map[np.where(label[i,:,:] == label_ids.index(_id))] = weights[label_ids.index(_id)]
                edge = np.float32(morph.binary_dilation(
                    canny(np.float32(label[i,:,:] == label_ids.index(_id)), sigma=1), selem=selem))
                edge_frequency = np.sum(np.sum(edge == 1.0))
                if edge_frequency:
                    slice_map[np.where(edge == 1.0)] += edgescale * label[i,:,:].size / edge_frequency
        weight_map = np.append(weight_map, np.expand_dims(slice_map, axis=0), axis=0)
    return np.float32(weight_map)

# ********************** Data Preprocessing Functions *******************#
def preprocess_volume(volume, is_mask=False):
    """
    Preprocess a 3D volume: crop/pad to target size and normalize if it's an image
    """
    target_shape = ACDC_SHAPE

    if len(volume.shape) != 3:
        raise ValueError(f"Expected 3D volume but got shape {volume.shape}")

    processed_channel = process_single_channel(volume, target_shape, is_mask)
    return np.expand_dims(processed_channel, axis=0)

def process_single_channel(channel_data, target_shape, is_mask=False):
    """
    Process a single channel of a 3D volume
    """
    depth, height, width = channel_data.shape
    target_d, target_h, target_w = target_shape

    # Depth processing
    if depth > target_d:
        start = (depth - target_d) // 2
        channel_data = channel_data[start:start+target_d, :, :]
    elif depth < target_d:
        pad_before = (target_d - depth) // 2
        pad_after = target_d - depth - pad_before
        channel_data = np.pad(channel_data, ((pad_before, pad_after), (0,0), (0,0)), mode='constant')

    # Height processing
    if height > target_h:
        start = (height - target_h) // 2
        channel_data = channel_data[:, start:start+target_h, :]
    elif height < target_h:
        pad_before = (target_h - height) // 2
        pad_after = target_h - height - pad_before
        channel_data = np.pad(channel_data, ((0,0), (pad_before, pad_after), (0,0)), mode='constant')

    # Width processing
    if width > target_w:
        start = (width - target_w) // 2
        channel_data = channel_data[:, :, start:start+target_w]
    elif width < target_w:
        pad_before = (target_w - width) // 2
        pad_after = target_w - width - pad_before
        channel_data = np.pad(channel_data, ((0,0), (0,0), (pad_before, pad_after)), mode='constant')

    # Normalize images (not masks)
    if not is_mask:
        channel_data = channel_data.astype(np.float32)
        if channel_data.max() > 0:
            channel_data /= channel_data.max()

    return channel_data

def normalize(image, scheme='zscore'):
    """Normalize image using different schemes"""
    if scheme == 'zscore':
        return normalize_zscore(image, z=0.5, offset=0)
    elif scheme == 'minmax':
        return normalize_minmax(image)
    elif scheme == 'truncated_zscore':
        return normalize_zscore(image, z=2, offset=0.5, clip=True)
    else:
        return image

def normalize_zscore(data, z=2, offset=0.5, clip=False):
    """Z-score normalization"""
    mean = np.mean(data)
    std = np.std(data)
    img = ((data - mean) / (2 * std * z) + offset)
    if clip:
        img = np.clip(img, -0.0, 1.0)
    return img

def normalize_minmax(data):
    """Min-max normalization"""
    _min = np.float(np.min(data))
    _max = np.float(np.max(data))
    if (_max - _min) != 0:
        img = (data - _min) / (_max - _min)
    else:
        img = np.zeros_like(data)
    return img

# ********************** Data Augmentation Functions *******************#
def sample_augmentation_parameters(transformation):
    """Sample random augmentation parameters"""
    if set(transformation.keys()) == {'patch_size', 'mm_patch_size'} or \
       set(transformation.keys()) == {'patch_size', 'mm_patch_size', 'mask_roi'}:
        return None

    shift_x = rng.uniform(*transformation.get('translation_range_x', [0., 0.]))
    shift_y = rng.uniform(*transformation.get('translation_range_y', [0., 0.]))
    translation = (shift_x, shift_y)
    rotation = rng.uniform(*transformation.get('rotation_range', [0., 0.]))
    shear = rng.uniform(*transformation.get('shear_range', [0., 0.]))
    roi_scale = rng.uniform(*transformation.get('roi_scale_range', [1., 1.]))
    z = rng.uniform(*transformation.get('zoom_range', [1., 1.]))
    zoom = (z, z)

    if 'do_flip' in transformation:
        if type(transformation['do_flip']) == tuple:
            flip_x = rng.randint(2) > 0 if transformation['do_flip'][0] else False
            flip_y = rng.randint(2) > 0 if transformation['do_flip'][1] else False
        else:
            flip_x = rng.randint(2) > 0 if transformation['do_flip'] else False
            flip_y = False
    else:
        flip_x, flip_y = False, False

    sequence_shift = rng.randint(30) if transformation.get('sequence_shift', False) else 0

    return namedtuple('Params', ['translation', 'rotation', 'shear', 'zoom',
                                 'roi_scale', 'flip_x', 'flip_y',
                                 'sequence_shift'])(translation, rotation, shear, zoom,
                                                    roi_scale, flip_x, flip_y, sequence_shift)

def build_augmentation_transform(rotation=0, shear=0, translation=(0, 0), flip_x=False, flip_y=False, zoom=(1.0, 1.0)):
    """Build augmentation transformation"""
    if flip_x:
        shear += 180
    if flip_y:
        shear += 180
        rotation += 180

    tform_augment = skimage.transform.AffineTransform(
        scale=(1. / zoom[0], 1. / zoom[1]),
        rotation=np.deg2rad(rotation),
        shear=np.deg2rad(shear),
        translation=translation
    )
    return tform_augment

def fast_warp(img, tf, output_shape, mode='constant', order=1):
    """Fast image warping using standard skimage.transform.warp"""
    try:
        # Use the standard warp function instead of _warps_cy
        return skimage.transform.warp(img, tf, output_shape=output_shape, mode=mode, order=order, preserve_range=True)
    except Exception as e:
        print(f"Warning: warp failed, using identity transform. Error: {e}")
        return img

def generate_noisy_image(noise_typ, image):
    """Generate noisy images for data augmentation"""
    if noise_typ == "gauss":
        row, col = image.shape[:2]
        mean = 0
        var = 0.0001
        sigma = var**0.5
        gauss = rng.normal(mean, sigma, (row, col))
        gauss = gauss.reshape(row, col, 1)
        noisy = image + gauss
        return noisy
    elif noise_typ == "s&p":
        s_vs_p = 0.5
        amount = 0.004
        out = np.copy(image)
        # Salt mode
        num_salt = np.ceil(amount * image.size * s_vs_p)
        coords = [rng.randint(0, i - 1, int(num_salt)) for i in image.shape]
        out[coords] = 1
        # Pepper mode
        num_pepper = np.ceil(amount * image.size * (1. - s_vs_p))
        coords = [rng.randint(0, i - 1, int(num_pepper)) for i in image.shape]
        out[coords] = 0
        return out
    elif noise_typ == "poisson":
        vals = len(np.unique(image))
        vals = 2 ** np.ceil(np.log2(vals))
        noisy = rng.poisson(image * vals) / float(vals)
        return noisy
    elif noise_typ == "speckle":
        row, col = image.shape[:2]
        gauss = 0.1 * rng.randn(row, col)
        gauss = gauss.reshape(row, col, 1)
        noisy = image + image * gauss
        return noisy
    else:
        return image

def apply_augmentation_to_volume(volume, mask, augmentation_params, n_augmentations=3):
    """
    Apply data augmentation to 3D volumes
    Returns list of augmented (image, mask) pairs
    """
    augmented_data = []

    # Always include original
    augmented_data.append((volume.copy(), mask.copy()))

    for aug_idx in range(n_augmentations):
        try:
            # Sample augmentation parameters
            random_params = sample_augmentation_parameters(augmentation_params)

            if random_params is None:
                # If no augmentation params, just add the original again
                augmented_data.append((volume.copy(), mask.copy()))
                continue

            # Build transformation
            augment_tform = build_augmentation_transform(
                rotation=random_params.rotation,
                shear=random_params.shear,
                translation=random_params.translation,
                flip_x=random_params.flip_x,
                flip_y=random_params.flip_y,
                zoom=random_params.zoom
            )

            # Apply transformation to each slice in the volume
            aug_volume = np.zeros_like(volume)
            aug_mask = np.zeros_like(mask)

            for slice_idx in range(volume.shape[1]):  # Loop through depth dimension
                # Get 2D slice
                img_slice = volume[0, slice_idx, :, :]  # [H, W]
                mask_slice = mask[0, slice_idx, :, :]   # [H, W]

                # Apply transformation to image (order=1 for linear interpolation)
                aug_img_slice = fast_warp(img_slice, augment_tform,
                                        output_shape=img_slice.shape, mode='reflect', order=1)

                # Apply transformation to mask (order=0 for nearest neighbor)
                aug_mask_slice = fast_warp(mask_slice, augment_tform,
                                         output_shape=mask_slice.shape, mode='constant', order=0)

                # Store augmented slices
                aug_volume[0, slice_idx, :, :] = aug_img_slice
                aug_mask[0, slice_idx, :, :] = aug_mask_slice

            # Add noise if specified
            if 'add_noise' in augmentation_params and augmentation_params['add_noise']:
                noise_type = augmentation_params['add_noise'][rng.randint(len(augmentation_params['add_noise']))]
                if noise_type != 'none':
                    # Apply noise to each slice
                    for slice_idx in range(aug_volume.shape[1]):
                        slice_2d = aug_volume[0, slice_idx, :, :]
                        noisy_slice = generate_noisy_image(noise_type, slice_2d)
                        aug_volume[0, slice_idx, :, :] = noisy_slice

            augmented_data.append((aug_volume, aug_mask))
            print(f"  ✓ Generated augmentation {aug_idx + 1}/{n_augmentations}")

        except Exception as e:
            print(f"  ✗ Error in augmentation {aug_idx + 1}: {str(e)}")
            # If augmentation fails, add the original as fallback
            augmented_data.append((volume.copy(), mask.copy()))
            continue

    return augmented_data

def simple_augment_volume(volume, mask, n_augmentations=3):
    """
    Simple augmentation that always works - basic flips and rotations
    """
    augmented_data = []

    # Always include original
    augmented_data.append((volume.copy(), mask.copy()))

    augmentation_types = [
        ('flip_horizontal', lambda x: np.flip(x, axis=3)),  # Flip width
        ('flip_vertical', lambda x: np.flip(x, axis=2)),    # Flip height
        ('rotate_90', lambda x: np.rot90(x, k=1, axes=(2, 3))),
        ('rotate_180', lambda x: np.rot90(x, k=2, axes=(2, 3))),
        ('rotate_270', lambda x: np.rot90(x, k=3, axes=(2, 3))),
    ]

    for i in range(min(n_augmentations, len(augmentation_types))):
        try:
            aug_name, aug_func = augmentation_types[i]
            aug_volume = aug_func(volume)
            aug_mask = aug_func(mask)
            augmented_data.append((aug_volume, aug_mask))
            print(f"  ✓ Generated {aug_name} augmentation")
        except Exception as e:
            print(f"  ✗ Error in {aug_name}: {str(e)}")

    # If we need more augmentations, add some with small rotations using scipy
    remaining = n_augmentations - len(augmented_data) + 1
    for i in range(remaining):
        try:
            # Small random rotation
            angle = rng.uniform(-15, 15)
            aug_volume = snd.rotate(volume, angle, axes=(2, 3), reshape=False, mode='reflect', order=1)
            aug_mask = snd.rotate(mask, angle, axes=(2, 3), reshape=False, mode='constant', order=0)
            augmented_data.append((aug_volume, aug_mask))
            print(f"  ✓ Generated rotation ({angle:.1f}°) augmentation")
        except Exception as e:
            print(f"  ✗ Error in rotation augmentation: {str(e)}")

    return augmented_data

def resize_image_with_crop_or_pad_3D(image, target_height, target_width):
    """Crop and/or pad 3D image to target size"""
    height, width, _ = image.shape
    width_diff = target_width - width
    offset_crop_width = max(-width_diff // 2, 0)
    offset_pad_width = max(width_diff // 2, 0)

    height_diff = target_height - height
    offset_crop_height = max(-height_diff // 2, 0)
    offset_pad_height = max(height_diff // 2, 0)

    # Crop if needed
    cropped = crop_to_bounding_box_3D(image, offset_crop_height, offset_crop_width,
                                     min(target_height, height), min(target_width, width))

    # Pad if needed
    resized = pad_to_bounding_box_3D(cropped, offset_pad_height, offset_pad_width,
                                    target_height, target_width)
    return resized

def pad_to_bounding_box_3D(image, offset_height, offset_width, target_height, target_width):
    """Pad 3D image to bounding box"""
    height, width, _ = image.shape
    after_padding_width = target_width - offset_width - width
    after_padding_height = target_height - offset_height - height

    assert offset_height >= 0, "offset_height must be >= 0"
    assert offset_width >= 0, "offset_width must be >= 0"
    assert after_padding_width >= 0, "after_padding_width must be >= 0"
    assert after_padding_height >= 0, "after_padding_height must be >= 0"

    padded = np.lib.pad(image,
                       ((offset_height, after_padding_height),
                        (offset_width, after_padding_width),
                        (0, 0)), 'constant')
    return padded

def crop_to_bounding_box_3D(image, offset_height, offset_width, target_height, target_width):
    """Crop 3D image to bounding box"""
    height, width, _ = image.shape

    assert offset_width >= 0, "offset_width must be >= 0"
    assert offset_height >= 0, "offset_height must be >= 0"
    assert target_width > 0, "target_width must be > 0"
    assert target_height > 0, "target_height must be > 0"
    assert width >= (target_width + offset_width), "width must be >= target + offset"
    assert height >= (target_height + offset_height), "height must be >= target + offset"

    cropped = image[offset_height: target_height+offset_height,
                   offset_width: target_width+offset_width, :]
    return cropped

# ********************** Enhanced Preprocessing with Augmentation *******************#
def enhanced_preprocess_acdc_data(data_path, output_dir, augmentation_params=None, augment_per_file=3, use_simple_aug=True):
    """
    Enhanced preprocessing with data augmentation for ACDC dataset
    """
    mkdir(output_dir)

    # Default augmentation parameters
    if augmentation_params is None:
        augmentation_params = {
            'train': {
                'patch_size': (224, 224),
                'translation_range_x': [-10, 10],
                'translation_range_y': [-10, 10],
                'rotation_range': [-15, 15],
                'zoom_range': [0.9, 1.1],
                'do_flip': (True, False),
                'add_noise': ['gauss', 's&p', 'none']
            },
            'val': {
                'patch_size': (224, 224)
            },
            'test': {
                'patch_size': (224, 224)
            }
        }

    # Find all HDF5 files
    h5_files = sorted(glob(os.path.join(data_path, '**/*.h5'), recursive=True))
    print(f"Found {len(h5_files)} HDF5 files")

    processed_count = 0
    augmented_count = 0

    for i, h5_file in enumerate(h5_files):
        try:
            print(f"Processing {i+1}/{len(h5_files)}: {os.path.basename(h5_file)}")

            with h5py.File(h5_file, 'r') as f:
                # Extract image
                if 'img' in f:
                    image_data = np.array(f['img'])
                elif 'data' in f:
                    image_data = np.array(f['data'])
                else:
                    for key in f.keys():
                        if isinstance(f[key], h5py.Dataset) and len(f[key].shape) >= 2:
                            image_data = np.array(f[key])
                            break
                    else:
                        print(f"No image data in {h5_file}")
                        continue

                # Extract segmentation
                seg_data = None
                if 'seg' in f:
                    seg_data = np.array(f['seg'])
                elif 'label' in f:
                    seg_data = np.array(f['label'])

                # Handle 4D (time, depth, H, W) or 3D (depth, H, W)
                if len(image_data.shape) == 4:
                    image_volume = image_data[0]  # take first time frame (ED)
                    seg_volume = seg_data[0] if seg_data is not None else None
                elif len(image_data.shape) == 3:
                    image_volume = image_data
                    seg_volume = seg_data
                else:
                    print(f"Unexpected shape: {image_data.shape}")
                    continue

                # Remove extra channel dimension if present
                if len(image_volume.shape) == 4:
                    image_volume = image_volume[0]
                if seg_volume is not None and len(seg_volume.shape) == 4:
                    seg_volume = seg_volume[0]

                # Basic preprocessing
                processed_image = preprocess_volume(image_volume, is_mask=False)
                if seg_volume is not None:
                    processed_mask = preprocess_volume(seg_volume, is_mask=True)
                else:
                    print(f"No segmentation mask in {h5_file}")
                    continue

                file_id = os.path.splitext(os.path.basename(h5_file))[0]

                # Save original processed data
                np.save(os.path.join(output_dir, f'img_{file_id}_orig.npy'), processed_image)
                np.save(os.path.join(output_dir, f'mask_{file_id}_orig.npy'), processed_mask)
                processed_count += 1

                # Apply data augmentation
                print(f"  Generating {augment_per_file} augmentations...")

                if use_simple_aug:
                    # Use simple augmentation that always works
                    augmented_pairs = simple_augment_volume(
                        processed_image, processed_mask,
                        n_augmentations=augment_per_file
                    )
                else:
                    # Try advanced augmentation
                    augmented_pairs = apply_augmentation_to_volume(
                        processed_image, processed_mask,
                        augmentation_params['train'],
                        n_augmentations=augment_per_file
                    )

                # Save augmented data (skip the first one which is original)
                for aug_idx, (aug_img, aug_mask) in enumerate(augmented_pairs[1:], 1):
                    np.save(os.path.join(output_dir, f'img_{file_id}_aug{aug_idx}.npy'), aug_img)
                    np.save(os.path.join(output_dir, f'mask_{file_id}_aug{aug_idx}.npy'), aug_mask)
                    augmented_count += 1

                successful_augs = len(augmented_pairs) - 1
                print(f"  ✓ Saved original + {successful_augs} augmented versions for {file_id}")

        except Exception as e:
            print(f"Error processing {h5_file}: {str(e)}")
            import traceback
            traceback.print_exc()

    print(f"\n📊 Processing Summary:")
    print(f"  Original files processed: {processed_count}")
    print(f"  Augmented files created: {augmented_count}")
    print(f"  Total files: {processed_count + augmented_count}")

    return processed_count, augmented_count

# ********************** Data Splitting with Counting *******************#
def split_data_with_count(data_dir, test_size=0.2, val_size=0.1, random_state=42):
    """Split processed data into train/val/test directories and count the splits"""
    img_files = sorted(glob(os.path.join(data_dir, 'img_*.npy')))
    mask_files = []

    for img_file in img_files:
        file_id = os.path.basename(img_file).replace('img_', '').replace('.npy', '')
        mask_file = os.path.join(data_dir, f'mask_{file_id}.npy')
        if os.path.exists(mask_file):
            mask_files.append(mask_file)
        else:
            print(f"Warning: Mask not found for {file_id}")
            mask_files.append(None)

    valid_samples = [(img, mask) for img, mask in zip(img_files, mask_files) if mask is not None]
    if not valid_samples:
        raise ValueError("No valid image-mask pairs found!")

    valid_img_files, valid_mask_files = zip(*valid_samples)

    # First split: separate test set
    train_val_imgs, test_imgs, train_val_masks, test_masks = train_test_split(
        valid_img_files, valid_mask_files, test_size=test_size, random_state=random_state
    )

    # Second split: separate validation from train
    val_size_adj = val_size / (1 - test_size)
    train_imgs, val_imgs, train_masks, val_masks = train_test_split(
        train_val_imgs, train_val_masks, test_size=val_size_adj, random_state=random_state
    )

    splits = {
        'train': (train_imgs, train_masks),
        'val': (val_imgs, val_masks),
        'test': (test_imgs, test_masks)
    }

    # Create directories and move files
    for split_name, (imgs, masks) in splits.items():
        split_dir = mkdir(os.path.join(data_dir, split_name))
        for img_file, mask_file in zip(imgs, masks):
            shutil.move(img_file, os.path.join(split_dir, os.path.basename(img_file)))
            shutil.move(mask_file, os.path.join(split_dir, os.path.basename(mask_file)))

    # Count and display results
    print_counts_by_split(data_dir)

    return get_split_counts(data_dir)

def print_counts_by_split(data_dir):
    """Print counts for each split"""
    print("\n📊 Dataset Split Counts:")
    print("=" * 50)

    for split_name in ['train', 'val', 'test']:
        split_dir = os.path.join(data_dir, split_name)
        if os.path.exists(split_dir):
            img_files = glob(os.path.join(split_dir, 'img_*.npy'))
            mask_files = glob(os.path.join(split_dir, 'mask_*.npy'))

            # Count original vs augmented
            orig_imgs = [f for f in img_files if '_orig' in f]
            aug_imgs = [f for f in img_files if '_aug' in f]

            print(f"{split_name.upper():<6}: {len(img_files):>3} images ({len(orig_imgs)} orig, {len(aug_imgs)} aug), "
                  f"{len(mask_files):>3} masks")
        else:
            print(f"{split_name.upper():<6}: Directory not found")

    total_imgs, total_masks = count_current_splits(data_dir)
    print("=" * 50)
    print(f"TOTAL : {total_imgs:>3} images, {total_masks:>3} masks")

def get_split_counts(data_dir):
    """Get counts for each split as dictionary"""
    counts = {}
    for split_name in ['train', 'val', 'test']:
        split_dir = os.path.join(data_dir, split_name)
        if os.path.exists(split_dir):
            img_count = len(glob(os.path.join(split_dir, 'img_*.npy')))
            counts[split_name] = img_count
        else:
            counts[split_name] = 0
    return counts

def count_current_splits(data_dir):
    """Count current split distribution"""
    total_images = 0
    total_masks = 0

    for split_name in ['train', 'val', 'test']:
        split_dir = os.path.join(data_dir, split_name)
        if os.path.exists(split_dir):
            img_count = len(glob(os.path.join(split_dir, 'img_*.npy')))
            mask_count = len(glob(os.path.join(split_dir, 'mask_*.npy')))

            total_images += img_count
            total_masks += mask_count

    return total_images, total_masks

# ********************** Main Execution *******************#
if __name__ == "__main__":
    # Configuration
    raw_data_path = "/content/ACDC_preprocessed"  # Change this to your ACDC data path
    processed_data_dir = "data/ACDC_processed_enhanced"

    # Augmentation parameters
    augmentation_config = {
        'train': {
            'patch_size': (224, 224),
            'translation_range_x': [-20, 20],
            'translation_range_y': [-20, 20],
            'rotation_range': [-25, 25],
            'zoom_range': [0.8, 1.3],
            'do_flip': (True, True),
            'add_noise': ['gauss', 's&p', 'poisson', 'speckle', 'none'],
            'data_augmentation': True
        },
        'val': {
            'patch_size': (224, 224),
            'data_augmentation': False
        },
        'test': {
            'patch_size': (224, 224),
            'data_augmentation': False
        }
    }

    print("🚀 Starting Enhanced ACDC Dataset Preprocessing with Augmentation...")

    # Process the data with augmentation (3 augmentations per file)
    # Using simple_aug=True for reliable augmentations
    original_count, augmented_count = enhanced_preprocess_acdc_data(
        raw_data_path,
        processed_data_dir,
        augmentation_config,
        augment_per_file=3,  # Generate 3 augmented versions per original
        use_simple_aug=True  # Use reliable simple augmentations
    )

    # Split the data and get counts
    split_counts = split_data_with_count(
        processed_data_dir,
        test_size=0.2,
        val_size=0.1,
        random_state=42
    )

    # Final detailed count
    print("\n" + "🔍 Final Detailed Count:".center(50))
    print_counts_by_split(processed_data_dir)

    print(f"\n✅ Enhanced ACDC dataset preprocessing completed!")
    print(f"📈 Final distribution: Train={split_counts['train']}, Val={split_counts['val']}, Test={split_counts['test']}")
    print(f"📊 Total files: {original_count + augmented_count} ({original_count} original + {augmented_count} augmented)")
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np
import os
from glob import glob

class ACDCDataset(Dataset):
    """
    PyTorch Dataset for 3D ACDC .npy files.
    Assumes files are saved as: (C, D, H, W) format from the preprocessing script
    """
    def __init__(self, root_dir, split="train", normalize=True, transform=None):
        """
        Args:
            root_dir: Root directory containing 'train', 'val', or 'test' subfolders
            split: 'train', 'val', or 'test'
            normalize: Whether to normalize images to [0, 1]
            transform: Optional transform to apply
        """
        self.root_dir = os.path.join(root_dir, split)
        self.split = split
        self.normalize = normalize
        self.transform = transform

        # Check if directory exists
        if not os.path.exists(self.root_dir):
            raise FileNotFoundError(f"Directory not found: {self.root_dir}")

        # Get all image and mask files
        self.img_files = sorted(glob(os.path.join(self.root_dir, "img_*.npy")))
        self.mask_files = sorted(glob(os.path.join(self.root_dir, "mask_*.npy")))

        # Verify matching files
        assert len(self.img_files) == len(self.mask_files), \
            f"Mismatch: {len(self.img_files)} images vs {len(self.mask_files)} masks"
        assert len(self.img_files) > 0, f"No image files found in {self.root_dir}"

        # Load first sample to get shape info
        sample_img = np.load(self.img_files[0], allow_pickle=False)
        sample_mask = np.load(self.mask_files[0], allow_pickle=False)

        # Expected shape from preprocessing: (1, D, H, W)
        if len(sample_img.shape) == 4:
            self.c, self.d, self.h, self.w = sample_img.shape
        else:
            raise ValueError(f"Expected 4D image (C,D,H,W), got {sample_img.shape}")

        if len(sample_mask.shape) == 4:
            self.mask_c, self.mask_d, self.mask_h, self.mask_w = sample_mask.shape
        else:
            raise ValueError(f"Expected 4D mask (C,D,H,W), got {sample_mask.shape}")

        print(f"Dataset initialized: {split} split, {len(self.img_files)} samples, "
              f"shape: ({self.c}, {self.d}, {self.h}, {self.w})")

    def __len__(self):
        return len(self.img_files)

    def __getitem__(self, idx):
        # Load image and mask
        img = np.load(self.img_files[idx], allow_pickle=False)
        mask = np.load(self.mask_files[idx], allow_pickle=False)

        # Convert to float32
        img = img.astype(np.float32)
        mask = mask.astype(np.int64)

        # Apply normalization if requested
        if self.normalize:
            # Normalize to [0, 1] if not already
            if img.max() > 1.0:
                img = img / 255.0

        # Remove channel dimension from mask if it's 1
        if mask.shape[0] == 1:
            mask = mask[0]  # Now shape: (D, H, W)

        # Convert to PyTorch tensors
        img_tensor = torch.from_numpy(img)  # (C, D, H, W)
        mask_tensor = torch.from_numpy(mask)  # (D, H, W)

        # Apply transforms if any
        if self.transform:
            img_tensor, mask_tensor = self.transform(img_tensor, mask_tensor)

        return img_tensor, mask_tensor


def create_dataloaders(root_dir, batch_size=2, normalize=True, num_workers=4,
                       train_split='train', val_split='val', test_split='test',
                       return_test=False):
    """
    Create train, validation, and test dataloaders

    Args:
        root_dir: Root directory containing train/val/test folders
        batch_size: Batch size
        normalize: Whether to normalize to [0, 1]
        num_workers: Number of workers for data loading
        train_split: Name of training folder
        val_split: Name of validation folder
        test_split: Name of test folder
        return_test: Whether to return test loader as well

    Returns:
        train_loader, val_loader [, test_loader]
    """
    # Create datasets
    train_dataset = ACDCDataset(
        root_dir=root_dir,
        split=train_split,
        normalize=normalize
    )

    val_dataset = ACDCDataset(
        root_dir=root_dir,
        split=val_split,
        normalize=normalize
    )

    # Create dataloaders
    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        num_workers=num_workers,
        pin_memory=True,
        drop_last=True
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=num_workers,
        pin_memory=True,
        drop_last=False
    )

    print("\n" + "="*60)
    print("DATALOADERS SUMMARY")
    print("="*60)
    print(f"Train loader: {len(train_loader)} batches, {len(train_dataset)} samples")
    print(f"Val loader: {len(val_loader)} batches, {len(val_dataset)} samples")

    if return_test:
        test_dataset = ACDCDataset(
            root_dir=root_dir,
            split=test_split,
            normalize=normalize
        )

        test_loader = DataLoader(
            test_dataset,
            batch_size=batch_size,
            shuffle=False,
            num_workers=num_workers,
            pin_memory=True,
            drop_last=False
        )

        print(f"Test loader: {len(test_loader)} batches, {len(test_dataset)} samples")
        print("="*60 + "\n")

        return train_loader, val_loader, test_loader

    print("="*60 + "\n")
    return train_loader, val_loader


def create_single_loader(root_dir, split='test', batch_size=2,
                         normalize=True, num_workers=4, shuffle=False):
    """
    Create a single dataloader for a specific split

    Args:
        root_dir: Root directory containing split folders
        split: 'train', 'val', or 'test'
        batch_size: Batch size
        normalize: Whether to normalize to [0, 1]
        num_workers: Number of workers for data loading
        shuffle: Whether to shuffle the data

    Returns:
        dataloader
    """
    dataset = ACDCDataset(
        root_dir=root_dir,
        split=split,
        normalize=normalize
    )

    loader = DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=shuffle,
        num_workers=num_workers,
        pin_memory=True,
        drop_last=False
    )

    print(f"\nCreated {split} loader: {len(loader)} batches, {len(dataset)} samples")
    return loader


def verify_dataset(root_dir):
    """
    Verify the integrity of the dataset
    """
    print("\n" + "="*60)
    print("DATASET VERIFICATION")
    print("="*60)

    all_valid = True

    for split in ['train', 'val', 'test']:
        split_dir = os.path.join(root_dir, split)
        if not os.path.exists(split_dir):
            print(f"⚠  {split} directory not found: {split_dir}")
            continue

        img_files = sorted(glob(os.path.join(split_dir, "img_*.npy")))
        mask_files = sorted(glob(os.path.join(split_dir, "mask_*.npy")))

        print(f"\n{split.upper()} Split:")
        print(f"  Images: {len(img_files)}")
        print(f"  Masks: {len(mask_files)}")

        if len(img_files) == 0:
            print(f"  ❌ No files found")
            all_valid = False
            continue

        # Check first sample
        try:
            img = np.load(img_files[0], allow_pickle=False)
            mask = np.load(mask_files[0], allow_pickle=False)

            print(f"  Sample image shape: {img.shape}, dtype: {img.dtype}")
            print(f"  Sample mask shape: {mask.shape}, dtype: {mask.dtype}")
            print(f"  Image range: [{img.min():.3f}, {img.max():.3f}]")
            print(f"  Unique mask values: {np.unique(mask)}")

            # Verify expected format (C,D,H,W) with C=1
            if len(img.shape) != 4:
                print(f"  ❌ Expected 4D image, got {len(img.shape)}D")
                all_valid = False
            elif img.shape[0] != 1:
                print(f"  ❌ Expected 1 channel, got {img.shape[0]}")
                all_valid = False
            else:
                print(f"  ✅ Format correct: (C={img.shape[0]}, D={img.shape[1]}, H={img.shape[2]}, W={img.shape[3]})")

        except Exception as e:
            print(f"  ❌ Error loading sample: {str(e)}")
            all_valid = False

    print("\n" + "="*60)
    if all_valid:
        print("✅ Dataset verification passed!")
    else:
        print("⚠  Dataset verification found issues!")
    print("="*60)

    return all_valid


if __name__ == "__main__":
    import argparse

    parser = argparse.ArgumentParser(description='ACDC Dataset Loader Test')
    parser.add_argument('--data_dir', type=str, default="data/ACDC_processed_enhanced",
                        help='Path to processed ACDC data')
    parser.add_argument('--batch_size', type=int, default=2,
                        help='Batch size for dataloaders')
    parser.add_argument('--num_workers', type=int, default=4,
                        help='Number of workers for data loading')
    parser.add_argument('--verify', action='store_true',
                        help='Run dataset verification')
    args = parser.parse_args([]) # <-- Changed this line

    # Verify dataset if requested
    if args.verify:
        verify_dataset(args.data_dir)

    # Create dataloaders with test set
    print("\n📦 Creating dataloaders...")
    train_loader, val_loader, test_loader = create_dataloaders(
        root_dir=args.data_dir,
        batch_size=args.batch_size,
        normalize=True,
        num_workers=args.num_workers,
        return_test=True
    )

    # Test a sample batch from each split
    print("\n📊 SAMPLE BATCHES:")
    print("-" * 60)

    # Training batch
    for images, masks in train_loader:
        print(f"Train batch:")
        print(f"  Images: {images.shape}  # [B, C, D, H, W]")
        print(f"  Masks:  {masks.shape}   # [B, D, H, W]")
        print(f"  Image dtype: {images.dtype}")
        print(f"  Mask dtype: {masks.dtype}")
        print(f"  Image range: [{images.min():.3f}, {images.max():.3f}]")
        print(f"  Unique mask values: {torch.unique(masks)}")
        break

    # Validation batch
    for images, masks in val_loader:
        print(f"\nValidation batch:")
        print(f"  Images: {images.shape}")
        print(f"  Masks:  {masks.shape}")
        break

    # Test batch
    for images, masks in test_loader:
        print(f"\nTest batch:")
        print(f"  Images: {images.shape}")
        print(f"  Masks:  {masks.shape}")
        break

    print("\n" + "="*60)
    print("✅ All dataloaders created successfully!")
    print("="*60)

    # Example training loop structure
    """
    # Model configuration for ACDC
    model = get_umamba_bot_3d(
        num_input_channels=1,  # ACDC has 1 channel
        num_classes=4,         # ACDC has 4 classes (0: BG, 1: RV, 2: MYO, 3: LV)
        deep_supervision=True
    )
    model = model.cuda()

    # Training loop
    for epoch in range(num_epochs):
        model.train()
        for batch_idx, (images, masks) in enumerate(train_loader):
            images = images.cuda()
            masks = masks.cuda()
            outputs = model(images)
            loss = criterion(outputs, masks)
            # ... training code ...

        # Validation
        model.eval()
        with torch.no_grad():
            for images, masks in val_loader:
                images = images.cuda()
                masks = masks.cuda()
                outputs = model(images)
                # ... validation code ...

    # Testing after training
    model.eval()
    with torch.no_grad():
        for images, masks in test_loader:
            images = images.cuda()
            masks = masks.cuda()
            outputs = model(images)
            # Calculate test metrics
    """
```

```python
import numpy as np
import math
import torch
from torch import nn
from torch.nn import functional as F
from typing import Union, Type, List, Tuple
from mamba_ssm import Mamba

# --- 1. Helper Layers ---

class UpsampleLayer(nn.Module):
    def __init__(self, conv_op, input_channels, output_channels, pool_op_kernel_size, mode='trilinear'):
        super().__init__()
        self.conv = conv_op(input_channels, output_channels, kernel_size=1)
        self.pool_op_kernel_size = pool_op_kernel_size
        self.mode = mode

    def forward(self, x):
        x = F.interpolate(x, scale_factor=self.pool_op_kernel_size, mode=self.mode, align_corners=False)
        x = self.conv(x)
        return x

class MambaLayer(nn.Module):
    def __init__(self, dim, d_state=16, d_conv=4, expand=2):
        super().__init__()
        self.dim = dim
        self.norm = nn.LayerNorm(dim)
        self.mamba = Mamba(
            d_model=dim,
            d_state=d_state,
            d_conv=d_conv,
            expand=expand,
        )

    @torch.amp.autocast('cuda', enabled=False)
    def forward(self, x):
        if x.dtype == torch.float16:
            x = x.type(torch.float32)
        B, C = x.shape[:2]
        img_dims = x.shape[2:]
        n_tokens = x.shape[2:].numel()

        # Flatten spatial dims to sequence: (B, C, D, H, W) -> (B, L, C)
        x_flat = x.reshape(B, C, n_tokens).transpose(-1, -2)
        x_norm = self.norm(x_flat)
        x_mamba = self.mamba(x_norm)

        # Restore spatial dims: (B, L, C) -> (B, C, D, H, W)
        out = x_mamba.transpose(-1, -2).reshape(B, C, *img_dims)
        return out

# --- 2. Building Blocks (Modified for U-Mamba) ---

class BasicResBlock(nn.Module):
    def __init__(self, conv_op, input_channels, output_channels, norm_op, norm_op_kwargs,
                 kernel_size=3, padding=1, stride=1, use_1x1conv=False,
                 nonlin=nn.LeakyReLU, nonlin_kwargs={'inplace': True}, conv_bias=False):
        super().__init__()
        self.conv1 = conv_op(input_channels, output_channels, kernel_size, stride=stride, padding=padding, bias=conv_bias)
        self.norm1 = norm_op(output_channels, **norm_op_kwargs)
        self.act1 = nonlin(**nonlin_kwargs)
        self.conv2 = conv_op(output_channels, output_channels, kernel_size, padding=padding, bias=conv_bias)
        self.norm2 = norm_op(output_channels, **norm_op_kwargs)
        self.act2 = nonlin(**nonlin_kwargs)
        self.conv3 = conv_op(input_channels, output_channels, kernel_size=1, stride=stride, bias=conv_bias) if use_1x1conv else None

    def forward(self, x):
        y = self.act1(self.norm1(self.conv1(x)))
        y = self.norm2(self.conv2(y))
        if self.conv3: x = self.conv3(x)
        return self.act2(y + x)

class UMambaBlock(nn.Module):
    """
    As per Diagram: Residual Block followed by a Mamba Layer.
    """
    def __init__(self, conv_op, input_channels, output_channels, norm_op, norm_op_kwargs,
                 kernel_size=3, padding=1, stride=1, use_1x1conv=False,
                 nonlin=nn.LeakyReLU, nonlin_kwargs={'inplace': True}, conv_bias=False):
        super().__init__()
        self.res_block = BasicResBlock(conv_op, input_channels, output_channels, norm_op, norm_op_kwargs,
                                       kernel_size, padding, stride, use_1x1conv, nonlin, nonlin_kwargs, conv_bias)
        self.mamba = MambaLayer(dim=output_channels)

    def forward(self, x):
        return self.mamba(self.res_block(x))

# --- 3. U-Mamba Components ---

class UNetResEncoder(nn.Module):
    def __init__(self, input_channels, n_stages, features_per_stage, conv_op, kernel_sizes, strides,
                 n_blocks_per_stage, conv_bias, norm_op, norm_op_kwargs, nonlin, nonlin_kwargs):
        super().__init__()
        self.stages = nn.ModuleList()
        self.output_channels = features_per_stage
        self.strides = strides
        self.conv_op = conv_op
        self.norm_op = norm_op
        self.norm_op_kwargs = norm_op_kwargs
        self.nonlin = nonlin
        self.nonlin_kwargs = nonlin_kwargs
        self.conv_bias = conv_bias
        self.kernel_sizes = kernel_sizes
        self.conv_pad_sizes = [k // 2 if isinstance(k, int) else [i // 2 for i in k] for k in kernel_sizes]

        curr_channels = input_channels
        for s in range(n_stages):
            # Each stage starts with a UMambaBlock (Res + Mamba) that handles downsampling
            stage = nn.Sequential(
                UMambaBlock(conv_op, curr_channels, features_per_stage[s], norm_op, norm_op_kwargs,
                            kernel_sizes[s], self.conv_pad_sizes[s], strides[s], use_1x1conv=True,
                            nonlin=nonlin, nonlin_kwargs=nonlin_kwargs, conv_bias=conv_bias),
                *[BasicResBlock(conv_op, features_per_stage[s], features_per_stage[s], norm_op, norm_op_kwargs,
                                kernel_sizes[s], self.conv_pad_sizes[s], 1, False, nonlin, nonlin_kwargs, conv_bias)
                  for _ in range(n_blocks_per_stage[s] - 1)]
            )
            self.stages.append(stage)
            curr_channels = features_per_stage[s]

    def forward(self, x):
        skips = []
        for stage in self.stages:
            x = stage(x)
            skips.append(x)
        return skips

class UNetResDecoder(nn.Module):
    def __init__(self, encoder, num_classes, n_conv_per_stage, deep_supervision):
        super().__init__()
        self.deep_supervision = deep_supervision
        self.stages = nn.ModuleList()
        self.upsample_layers = nn.ModuleList()
        self.seg_layers = nn.ModuleList()

        n_stages = len(encoder.output_channels)
        for s in range(1, n_stages):
            feat_below = encoder.output_channels[-s]
            feat_skip = encoder.output_channels[-(s + 1)]

            self.upsample_layers.append(UpsampleLayer(encoder.conv_op, feat_below, feat_skip, encoder.strides[-s]))

            # Decoder blocks are also UMambaBlocks
            self.stages.append(nn.Sequential(
                UMambaBlock(encoder.conv_op, 2 * feat_skip, feat_skip, encoder.norm_op, encoder.norm_op_kwargs,
                            encoder.kernel_sizes[-(s + 1)], encoder.conv_pad_sizes[-(s + 1)], 1, True,
                            encoder.nonlin, encoder.nonlin_kwargs, encoder.conv_bias),
                *[BasicResBlock(encoder.conv_op, feat_skip, feat_skip, encoder.norm_op, encoder.norm_op_kwargs,
                                encoder.kernel_sizes[-(s + 1)], encoder.conv_pad_sizes[-(s + 1)], 1, False,
                                encoder.nonlin, encoder.nonlin_kwargs, encoder.conv_bias)
                  for _ in range(n_conv_per_stage[s-1] - 1)]
            ))
            self.seg_layers.append(encoder.conv_op(feat_skip, num_classes, 1))

    def forward(self, skips):
        x = skips[-1]
        seg_outputs = []
        for s in range(len(self.stages)):
            x = self.upsample_layers[s](x)
            skip = skips[-(s+2)]
            if x.shape[2:] != skip.shape[2:]:
                x = F.interpolate(x, size=skip.shape[2:], mode='trilinear', align_corners=False)
            x = torch.cat([x, skip], dim=1)
            x = self.stages[s](x)
            if self.deep_supervision or s == (len(self.stages) - 1):
                seg_outputs.append(self.seg_layers[s](x))

        return seg_outputs[::-1] if self.deep_supervision else seg_outputs[-1]

# --- 4. Main Model Wrapper ---

class UMamba(nn.Module):
    def __init__(self, input_channels, n_stages, features_per_stage, num_classes, strides, n_conv_per_stage,
                 deep_supervision=True, conv_bias=True):
        super().__init__()
        self.encoder = UNetResEncoder(input_channels, n_stages, features_per_stage, nn.Conv3d, [3]*n_stages,
                                      strides, n_conv_per_stage, conv_bias, nn.InstanceNorm3d,
                                      {'eps': 1e-5, 'affine': True}, nn.LeakyReLU, {'inplace': True})
        self.decoder = UNetResDecoder(self.encoder, num_classes, n_conv_per_stage[:-1], deep_supervision)

    def forward(self, x):
        skips = self.encoder(x)
        return self.decoder(skips)

# --- 5. Factory Function ---

def get_umamba_3d(num_input_channels=1, num_classes=2):
    model = UMamba(
        input_channels=num_input_channels,
        num_classes=num_classes,
        n_stages=5,
        features_per_stage=(32, 64, 128, 256, 512),
        strides=(1, 2, 2, 2, 2),
        n_conv_per_stage=(2, 2, 2, 2, 2)
    )
    return model
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.amp import autocast, GradScaler
import numpy as np
from tqdm import tqdm
import os
import matplotlib.pyplot as plt
from typing import Optional, Tuple, List, Union
import gc

# Memory optimization
if torch.cuda.is_available():
    torch.cuda.empty_cache()
    gc.collect()

# --- First, let's debug the model architecture ---
def debug_model_architecture(model):
    """Print detailed model architecture and parameter counts"""
    print("\n" + "="*60)
    print("MODEL ARCHITECTURE DEBUG")
    print("="*60)

    total_params = 0
    print(f"{'Layer':<30} {'Output Shape':<20} {'Params':<10}")
    print("-"*60)

    for name, module in model.named_modules():
        if len(list(module.children())) == 0:  # Only leaf modules
            params = sum(p.numel() for p in module.parameters())
            total_params += params
            if params > 0:  # Only show modules with parameters
                print(f"{name[:30]:<30} {str(tuple(module.weight.shape) if hasattr(module, 'weight') else 'N/A'):<20} {params:<10,}")

    print("-"*60)
    print(f"{'TOTAL':<30} {'':<20} {total_params:<10,}")

    # Check Mamba layers specifically
    print("\nMAMBA LAYER DETAILS:")
    for name, module in model.named_modules():
        if 'mamba' in name.lower() and hasattr(module, 'mamba'):
            print(f"{name}: dim={module.dim}")
            mamba_params = sum(p.numel() for p in module.mamba.parameters())
            print(f"  - Mamba parameters: {mamba_params:,}")

    return total_params

def train_epoch_mixed_precision(
    model: nn.Module,
    loader: torch.utils.data.DataLoader,
    optimizer: torch.optim.Optimizer,
    scaler: GradScaler,
    epoch: int,
    deep_supervision: bool = True,
    device: str = 'cuda',
    grad_accum_steps: int = 4
) -> Tuple[float, float]:
    """
    Train for one epoch with mixed precision and gradient accumulation
    """
    model.train()
    total_loss = 0
    total_correct = 0
    total_pixels = 0

    optimizer.zero_grad()

    progress_bar = tqdm(loader, desc=f'Epoch {epoch}')

    for batch_idx, (images, masks) in enumerate(progress_bar):
        images = images.to(device, non_blocking=True)
        masks = masks.to(device, non_blocking=True)

        # Mixed precision forward pass
        with autocast(device_type='cuda'):
            outputs = model(images)

            # Handle deep supervision outputs
            if isinstance(outputs, list):
                loss = 0
                batch_correct = 0
                batch_pixels = 0

                for i, output in enumerate(outputs):
                    # Get target mask at appropriate resolution
                    if output.shape[2:] != masks.shape[1:]:
                        target = F.interpolate(
                            masks.unsqueeze(1).float(),
                            size=output.shape[2:],
                            mode='nearest'
                        ).squeeze(1).long()
                    else:
                        target = masks

                    # Calculate loss with deep supervision weights
                    loss += F.cross_entropy(output, target) * (0.5 ** (len(outputs) - i - 1))

                    # Calculate accuracy only on final output
                    if i == 0:
                        pred_classes = output.argmax(dim=1)
                        if pred_classes.shape != masks.shape:
                            pred_classes = F.interpolate(
                                pred_classes.unsqueeze(1).float(),
                                size=masks.shape[1:],
                                mode='nearest'
                            ).squeeze(1).long()
                        batch_correct = (pred_classes == masks).sum().item()
                        batch_pixels = masks.numel()
            else:
                # Single output case
                if outputs.shape[2:] != masks.shape[1:]:
                    outputs = F.interpolate(outputs, size=masks.shape[1:], mode='trilinear', align_corners=False)
                loss = F.cross_entropy(outputs, masks)
                pred_classes = outputs.argmax(dim=1)
                batch_correct = (pred_classes == masks).sum().item()
                batch_pixels = masks.numel()

            # Normalize loss for gradient accumulation
            loss = loss / grad_accum_steps

        # Mixed precision backward pass
        scaler.scale(loss).backward()

        # Gradient accumulation
        if (batch_idx + 1) % grad_accum_steps == 0:
            # Gradient clipping
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()

        # Update metrics
        if isinstance(outputs, list):
            total_correct += batch_correct
            total_pixels += batch_pixels
        else:
            total_correct += batch_correct
            total_pixels += batch_pixels

        total_loss += loss.item() * grad_accum_steps

        # Free memory
        del images, masks, outputs
        if batch_idx % 10 == 0:
            torch.cuda.empty_cache()

        # Update progress bar
        avg_loss = total_loss / (batch_idx + 1)
        avg_acc = total_correct / total_pixels if total_pixels > 0 else 0
        progress_bar.set_postfix({
            'loss': f'{avg_loss:.4f}',
            'acc': f'{avg_acc:.4f}',
            'mem': f'{torch.cuda.memory_allocated()/1024**3:.1f}GB'
        })

    # Final optimizer step
    if (len(loader) % grad_accum_steps) != 0:
        scaler.step(optimizer)
        scaler.update()
        optimizer.zero_grad()

    return total_loss / len(loader), total_correct / total_pixels


def validate_epoch_mixed_precision(
    model: nn.Module,
    loader: torch.utils.data.DataLoader,
    deep_supervision: bool = True,
    device: str = 'cuda'
) -> Tuple[float, float]:
    """
    Validation with mixed precision and memory optimization
    """
    model.eval()
    total_loss = 0
    total_correct = 0
    total_pixels = 0

    with torch.no_grad():
        for images, masks in tqdm(loader, desc='Validation'):
            images = images.to(device, non_blocking=True)
            masks = masks.to(device, non_blocking=True)

            # Mixed precision forward pass
            with autocast(device_type='cuda'):
                outputs = model(images)

                # Handle deep supervision outputs - use final output only
                if isinstance(outputs, list):
                    final_output = outputs[0]

                    if final_output.shape[2:] != masks.shape[1:]:
                        final_output = F.interpolate(
                            final_output,
                            size=masks.shape[1:],
                            mode='trilinear',
                            align_corners=False
                        )

                    loss = F.cross_entropy(final_output, masks)
                    pred_classes = final_output.argmax(dim=1)
                    correct = (pred_classes == masks).sum().item()

                    del outputs
                else:
                    if outputs.shape[2:] != masks.shape[1:]:
                        outputs = F.interpolate(outputs, size=masks.shape[1:], mode='trilinear', align_corners=False)

                    loss = F.cross_entropy(outputs, masks)
                    pred_classes = outputs.argmax(dim=1)
                    correct = (pred_classes == masks).sum().item()

            total_loss += loss.item()
            total_correct += correct
            total_pixels += masks.numel()

            del images, masks, loss, pred_classes
            torch.cuda.empty_cache()

    return total_loss / len(loader), total_correct / total_pixels


def train_with_mixed_precision(
    model: nn.Module,
    train_loader: torch.utils.data.DataLoader,
    val_loader: torch.utils.data.DataLoader,
    optimizer: torch.optim.Optimizer,
    scheduler: Optional[torch.optim.lr_scheduler._LRScheduler] = None,
    num_epochs: int = 100,
    deep_supervision: bool = True,
    device: str = 'cuda',
    checkpoint_dir: str = 'checkpoints',
    grad_clip: Optional[float] = None,
    grad_accum_steps: int = 4
) -> dict:
    """
    Complete training loop with mixed precision and memory optimizations
    """
    os.makedirs(checkpoint_dir, exist_ok=True)

    scaler = GradScaler(device='cuda')

    history = {
        'train_loss': [],
        'train_acc': [],
        'val_loss': [],
        'val_acc': [],
        'lr': []
    }

    best_val_acc = 0
    best_val_loss = float('inf')

    print("\n" + "="*60)
    print("TRAINING CONFIGURATION")
    print("="*60)
    print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
    print(f"Trainable parameters: {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")
    print(f"Device: {device}")
    print(f"Deep supervision: {deep_supervision}")
    print(f"Gradient accumulation steps: {grad_accum_steps}")
    print(f"Effective batch size: {train_loader.batch_size * grad_accum_steps}")
    print(f"Number of epochs: {num_epochs}")
    print("="*60 + "\n")

    for epoch in range(num_epochs):
        train_loss, train_acc = train_epoch_mixed_precision(
            model=model,
            loader=train_loader,
            optimizer=optimizer,
            scaler=scaler,
            epoch=epoch + 1,
            deep_supervision=deep_supervision,
            device=device,
            grad_accum_steps=grad_accum_steps
        )

        val_loss, val_acc = validate_epoch_mixed_precision(
            model=model,
            loader=val_loader,
            deep_supervision=deep_supervision,
            device=device
        )

        current_lr = optimizer.param_groups[0]['lr']
        if scheduler is not None:
            if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
                scheduler.step(val_loss)
            else:
                scheduler.step()
            current_lr = optimizer.param_groups[0]['lr']

        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        history['lr'].append(current_lr)

        print(f"Epoch {epoch+1}/{num_epochs} - "
              f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} - "
              f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f} - "
              f"LR: {current_lr:.2e}")

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_acc': val_acc,
                'val_loss': val_loss,
            }, os.path.join(checkpoint_dir, 'best_model_acc.pth'))
            print(f"  -> New best model (acc) saved! Val Acc: {val_acc:.4f}")

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_acc': val_acc,
                'val_loss': val_loss,
            }, os.path.join(checkpoint_dir, 'best_model_loss.pth'))
            print(f"  -> New best model (loss) saved! Val Loss: {val_loss:.4f}")

        torch.cuda.empty_cache()
        gc.collect()

    print("-"*60)
    print(f"Training completed! Best validation accuracy: {best_val_acc:.4f}")

    return history


# Main training script with corrected model configuration
if __name__ == "__main__":
    # Memory optimizations
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        gc.collect()
        os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

    # Model configuration
    num_input_channels = 1
    num_classes = 4
    deep_supervision = True

    # Let's use a more reasonable configuration for 3D
    n_stages = 4
    features_per_stage = (16, 32, 64, 128)  # Increased from (4,8,16,32)
    pool_op_kernel_sizes = (1, 2, 2, 2)
    n_conv_per_stage = (2, 2, 2, 2)  # Increased from (1,1,1,1)

    print("\n" + "="*60)
    print("CREATING MODEL")
    print("="*60)
    print(f"Input channels: {num_input_channels}")
    print(f"Number of classes: {num_classes}")
    print(f"Number of stages: {n_stages}")
    print(f"Features per stage: {features_per_stage}")
    print(f"Strides: {pool_op_kernel_sizes}")
    print(f"Conv per stage: {n_conv_per_stage}")

    # Create model
    model = UMamba(
        input_channels=num_input_channels,
        num_classes=num_classes,
        deep_supervision=deep_supervision,
        n_stages=n_stages,
        features_per_stage=features_per_stage,
        strides=pool_op_kernel_sizes,
        n_conv_per_stage=n_conv_per_stage,
        conv_bias=False  # Set to False to save memory
    )

    # Move model to GPU
    model = model.to('cuda')

    # Debug model architecture
    total_params = debug_model_architecture(model)

    # Check if Mamba layers are properly initialized
    print("\n" + "="*60)
    print("CHECKING MAMBA LAYERS")
    print("="*60)
    for name, module in model.named_modules():
        if isinstance(module, MambaLayer):
            print(f"Found MambaLayer: {name}")
            print(f"  - dim: {module.dim}")
            print(f"  - mamba params: {sum(p.numel() for p in module.mamba.parameters()):,}")

    # Create dataloaders
    train_loader, val_loader = create_dataloaders(
        root_dir="data/ACDC_processed_enhanced",
        batch_size=1,
        normalize=True,
        num_workers=2
    )

    # Use AdamW with appropriate learning rate
    optimizer = torch.optim.AdamW(
        model.parameters(),
        lr=1e-4,  # Increased back to 1e-4
        weight_decay=1e-5,
        betas=(0.9, 0.999)
    )

    # Learning rate scheduler
    scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(
        optimizer,
        T_0=10,
        T_mult=2,
        eta_min=1e-6
    )

    # Train
    history = train_with_mixed_precision(
        model=model,
        train_loader=train_loader,
        val_loader=val_loader,
        optimizer=optimizer,
        scheduler=scheduler,
        num_epochs=10,
        deep_supervision=True,
        device='cuda',
        checkpoint_dir='umamba_bot_checkpoints',
        grad_accum_steps=4
    )

    # Plot training history
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.plot(history['train_loss'], label='Train')
    plt.plot(history['val_loss'], label='Validation')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Loss Curves')
    plt.grid(True)

    plt.subplot(1, 3, 2)
    plt.plot(history['train_acc'], label='Train')
    plt.plot(history['val_acc'], label='Validation')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title('Accuracy Curves')
    plt.grid(True)

    plt.subplot(1, 3, 3)
    plt.plot(history['lr'])
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.title('Learning Rate Schedule')
    plt.grid(True)
    plt.yscale('log')

    plt.tight_layout()
    plt.savefig('training_history.png', dpi=150, bbox_inches='tight')
    plt.show()

    print("\n" + "="*60)
    print("TRAINING COMPLETED")
    print("="*60)
```

```python
import numpy as np
import torch.nn.functional as F
from tqdm import tqdm
from torch.amp import autocast

def compute_dice(pred, target, num_classes=4, smooth=1e-6):
    """
    pred, target: tensors of shape (N, D, H, W) with class indices
    Returns per-class Dice and mean Dice.
    """
    dice_per_class = []
    for cls in range(num_classes):
        pred_cls = (pred == cls)
        target_cls = (target == cls)
        intersection = (pred_cls & target_cls).sum().float()
        union = pred_cls.sum().float() + target_cls.sum().float()
        dice = (2. * intersection + smooth) / (union + smooth)
        dice_per_class.append(dice.item())
    return dice_per_class, np.mean(dice_per_class)

def test_model(model, test_loader, device='cuda', num_classes=4, use_mixed_precision=True):
    model.eval()
    total_loss = 0
    total_correct = 0
    total_pixels = 0
    all_dice = []  # will store per-class dice for each batch

    with torch.no_grad():
        for images, masks in tqdm(test_loader, desc='Testing'):
            images = images.to(device)
            masks = masks.to(device)

            # Mixed precision forward (optional)
            if use_mixed_precision:
                with autocast(device_type='cuda'):
                    outputs = model(images)
            else:
                outputs = model(images)

            # If deep supervision, take the final output (highest resolution)
            if isinstance(outputs, list):
                outputs = outputs[0]

            # Resize output to match mask spatial dimensions if necessary
            if outputs.shape[2:] != masks.shape[1:]:
                outputs = F.interpolate(
                    outputs,
                    size=masks.shape[1:],
                    mode='trilinear',
                    align_corners=False
                )

            # Loss
            loss = F.cross_entropy(outputs, masks)
            total_loss += loss.item()

            # Accuracy
            pred_classes = outputs.argmax(dim=1)  # (B, D, H, W)
            correct = (pred_classes == masks).sum().item()
            total_correct += correct
            total_pixels += masks.numel()

            # Dice for this batch
            dice_per_class, _ = compute_dice(pred_classes, masks, num_classes)
            all_dice.append(dice_per_class)

    # Aggregate metrics
    avg_loss = total_loss / len(test_loader)
    avg_acc = total_correct / total_pixels
    avg_dice_per_class = np.mean(all_dice, axis=0).tolist()
    avg_mean_dice = np.mean([np.mean(d) for d in all_dice])

    # Print results
    print("\n" + "="*50)
    print("TEST SET EVALUATION")
    print("="*50)
    print(f"Loss:       {avg_loss:.4f}")
    print(f"Accuracy:   {avg_acc:.4f}")
    print(f"Mean Dice:  {avg_mean_dice:.4f}")
    for i, dice in enumerate(avg_dice_per_class):
        class_name = ['BG', 'RV', 'MYO', 'LV'][i]
        print(f"  Dice {class_name}: {dice:.4f}")
    print("="*50)

    return {
        'loss': avg_loss,
        'accuracy': avg_acc,
        'dice_per_class': avg_dice_per_class,
        'mean_dice': avg_mean_dice
    }
```

```python
import matplotlib.pyplot as plt

def visualize_predictions(model, test_loader, device, num_samples=4, save_path='test_predictions.png'):
    model.eval()
    fig, axes = plt.subplots(num_samples, 3, figsize=(12, 4*num_samples))
    with torch.no_grad():
        for i, (images, masks) in enumerate(test_loader):
            if i >= num_samples:
                break
            images = images.to(device)
            outputs = model(images)
            if isinstance(outputs, list):
                outputs = outputs[0]
            preds = outputs.argmax(dim=1).cpu().numpy()[0]   # first in batch
            # Take middle slice along depth
            slice_idx = images.shape[2] // 2
            img_slice = images[0, 0, slice_idx].cpu().numpy()
            mask_slice = masks[0, slice_idx].cpu().numpy()
            pred_slice = preds[slice_idx]

            axes[i, 0].imshow(img_slice, cmap='gray')
            axes[i, 0].set_title('Input')
            axes[i, 1].imshow(mask_slice, cmap='viridis', vmin=0, vmax=3)
            axes[i, 1].set_title('Ground Truth')
            axes[i, 2].imshow(pred_slice, cmap='viridis', vmin=0, vmax=3)
            axes[i, 2].set_title('Prediction')
    plt.tight_layout()
    plt.savefig(save_path)
    plt.show()

visualize_predictions(model, test_loader, device='cuda')
```
